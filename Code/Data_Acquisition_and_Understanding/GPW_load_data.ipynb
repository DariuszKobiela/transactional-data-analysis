```python
plt.plot(df['Timestamp?'][:100], df['?Cena'][:100])
plt.show()
```

```python
#df.loc[(df[â€˜Colorâ€™] == â€˜Greenâ€™) & (df[â€˜Shapeâ€™] == â€˜Rectangleâ€™)]
df_prepared = df.loc[(df['MIC - Market Identifier Code'] == 'WETP')]
df_prepared
```

```python
plt.plot(df_prepared['Timestamp?'], df_prepared['?Cena'])
plt.show()
```

```python
df_sorted = df_prepared.sort_values(by=['Timestamp?'], axis=0, ascending=True)
df_sorted
```

```python
plt.figure(figsize=(16,12))
plt.plot(df_sorted['Timestamp?'], df_sorted['?Cena'], color='orange', linewidth=1)
plt.title('WETP - WARSAW STOCK EXCHANGE/ ETPs during 2020-12-03')
plt.xlabel('Timestamps')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.bar(df_sorted['Timestamp?'], df_sorted['?Wolumen'], color='black', linewidth=1)
plt.title('WETP - WARSAW STOCK EXCHANGE Volume / ETPs during 2020-12-03')
plt.xlabel('Timestamps')
plt.ylabel('Volume')
plt.grid()
plt.show()
```
