## WORKING SCRIPT

```python
import pandas as pd
import glob

path = r'E:/1_Studia/PRACA_INŻYNIERSKA/NASDAQ_umowa/SAMPLE_TRADES_DATA' # use your path to trades file
all_files = glob.glob(path + "/*.csv")

list_of_values = []
names = ['index','seqnum','timestamp','mktcenter','price','shares','salescondition','canceled','dottchar','issuechar','msgseqnum','originalmsgseqnum','submkt']

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=None)
    list_of_values.append(df)

frame = pd.concat(list_of_values, axis=0, ignore_index=True, names=names)
frame.columns = names
frame = frame.drop('index', 1)
frame
```
