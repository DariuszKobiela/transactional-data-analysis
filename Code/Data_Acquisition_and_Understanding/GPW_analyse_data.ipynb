## WYKRESIKI

```python
df_prepared = df.loc[(df['MIC - Market Identifier Code'] == 'WETP')]
#df_prepared.head()
```

```python
df_sorted = df_prepared.sort_values(by=['Timestamp'], axis=0, ascending=True)
#df_sorted.head()
```

```python
%%time 
#czas wykonania komórki
# WYKRES CENY
plt.figure(figsize=(16,12))
plt.plot(df_sorted['Timestamp'], df_sorted['Kurs'], color='orange', linewidth=1)
plt.title('WETP - WARSAW STOCK EXCHANGE/ ETPs during 2020-12-03')
plt.xlabel('Timestamps')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
%%time 
#czas wykonania komórki
# WYKRES WOLUMENU
plt.figure(figsize=(16,12))
plt.bar(df_sorted['Timestamp'], df_sorted['Wolumen'], color='black', linewidth=1)
plt.title('WETP - WARSAW STOCK EXCHANGE Volume / ETPs during 2020-12-03')
plt.xlabel('Timestamps')
plt.ylabel('Volume')
plt.grid()
plt.show()
```
