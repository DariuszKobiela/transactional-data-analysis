```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('INTC/INTC_110801_110930.csv')
df[-100:-1]
```

```python
df.info()
```

```python
print(df.dtypes)
```

```python
df['<DATE>'] = pd.to_datetime(df['<DATE>'])
```

```python
df = df.set_index('<DATE>')
```

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight')

# Plot the time series in your DataFrame
ax = df.plot(color='blue')

# Specify the x-axis label in your plot
ax.set_xlabel('Date')

# Specify the y-axis label in your plot
ax.set_ylabel('Number of great discoveries')

ax.set_title('The title of my plot')

# Show plot
plt.show()
```

```python
ax = df['<LAST>'].plot(figsize=(12, 5), fontsize=12,linewidth=3, linestyle='--')
ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('Closing price values', fontsize=16)
ax.set_title('Intel prices during one month', fontsize=16)
plt.tight_layout()
plt.show()
```
